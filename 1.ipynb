```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import Dataset, DataLoader, random_split
import numpy as np
import os
import math
from torch.cuda.amp import autocast, GradScaler
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import torchvision.transforms as transforms
import torchvision.transforms as transforms

os.makedirs('./figures', exist_ok=True)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class HARDataset(Dataset):
    def __init__(self, data, labels, augment=False):
        self.data = data
        self.labels = labels
        self.label_map = {name: idx for idx, name in enumerate(sorted(set(labels)))}
        self.augment = augment

        self.transform = transforms.Compose([
            transforms.RandomHorizontalFlip(),
            transforms.RandomVerticalFlip(),
            transforms.RandomRotation(10),
            transforms.RandomApply([transforms.GaussianBlur(3)], p=0.5),
        ])
        
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = self.data[idx]
        label = self.label_map[self.labels[idx]]
        
        sample = (sample - sample.min()) / (sample.max() - sample.min() + 1e-8)
        sample = torch.tensor(sample, dtype=torch.float32).permute(1, 0, 2, 3)
        
        if self.augment:
            transformed = []
            for t in range(sample.size(1)):
                frame = sample[:, t, :, :].unsqueeze(1)
                transformed.append(self.transform(frame))
            sample = torch.cat(transformed, dim=1)
            
        return sample, label

# 时空注意力机制
class SpatioTemporalAttention(nn.Module):
    def __init__(self, input_dim, embed_dim=1024):
        super().__init__()
        self.temp_proj = nn.Linear(input_dim, embed_dim)
        self.temporal_attn = nn.MultiheadAttention(embed_dim, 4)
        self.spatial_proj = nn.Linear(input_dim, embed_dim)
        self.spatial_attn = nn.MultiheadAttention(embed_dim, 4)
        self.norm = nn.LayerNorm(embed_dim)
    
    def forward(self, x):
        B, T, H, W, C = x.shape
        # 时间注意力
        temp = x.reshape(B, T, -1)  # [B, T, H*W*C]
        temp = self.temp_proj(temp)  # 投影到embed_dim [B, T, embed_dim]
        temp = temp.permute(1, 0, 2)  # [T, B, embed_dim]
        temp, _ = self.temporal_attn(temp, temp, temp)
        
        # 空间注意力
        spatial = x.reshape(B, T, H * W * C)  # [B, T, H*W*C]
        spatial = self.spatial_proj(spatial)  # 投影到embed_dim [B, T, embed_dim]
        spatial = spatial.permute(1, 0, 2)  # [T, B, embed_dim]
        spatial, _ = self.spatial_attn(spatial, spatial, spatial)
        
        # 特征融合
        fused = self.norm(temp + spatial)
        return fused

# Transformer模块
class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, dropout=0.3):
        super().__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True)
        self.norm1 = nn.LayerNorm(embed_dim)
        self.ffn = nn.Sequential(
            nn.Linear(embed_dim, 4*embed_dim),
            nn.GELU(),
            nn.Linear(4*embed_dim, embed_dim),
            nn.Dropout(dropout)
        )
        self.norm2 = nn.LayerNorm(embed_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        x = self.norm1(x + self.dropout(attn_output))
        ffn_output = self.ffn(x)
        x = self.norm2(x + self.dropout(ffn_output))
        return x

# 优化的ResNet块
class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1):
        super().__init__()
        self.conv1 = nn.Conv3d(in_channels, out_channels, kernel_size=3, stride=stride, padding=1, bias=False)
        self.bn1 = nn.BatchNorm3d(out_channels)
        self.conv2 = nn.Conv3d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm3d(out_channels)
        self.relu = nn.GELU()
        
        self.shortcut = nn.Sequential()
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv3d(in_channels, out_channels, kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm3d(out_channels)
            )

    def forward(self, x):
        identity = self.shortcut(x)
        out = self.conv1(x)
        out = self.bn1(out)
        out = self.relu(out)
        out = self.conv2(out)
        out = self.bn2(out)
        out += identity
        return self.relu(out)

# 多尺度卷积
class MultiScaleConv3D(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        mid_channels = out_channels // 4
        
        self.branches = nn.ModuleList([
            nn.Conv3d(in_channels, mid_channels, 1),
            nn.Sequential(
                nn.Conv3d(in_channels, mid_channels, 3, padding=1),
                nn.GELU()
            ),
            nn.Sequential(
                nn.Conv3d(in_channels, mid_channels, 5, padding=2),
                nn.GELU()
            ),
            nn.Sequential(
                nn.MaxPool3d(3, stride=1, padding=1),
                nn.Conv3d(in_channels, mid_channels, 1)
            )
        ])
        
    def forward(self, x):
        return torch.cat([branch(x) for branch in self.branches], dim=1)

# 优化的HARnet模型
class HARnet(nn.Module):
    def __init__(self, num_classes=6):
        super().__init__()
        # 3D CNN backbone
        self.cnn = nn.Sequential(
            MultiScaleConv3D(10, 64),
            nn.BatchNorm3d(64),
            nn.GELU(),
            nn.MaxPool3d(3, stride=2, padding=1),
            ResidualBlock(64, 128, stride=2),
            ResidualBlock(128, 256, stride=2),
            MultiScaleConv3D(256, 512),
            nn.BatchNorm3d(512),
            nn.GELU(),
            nn.AdaptiveAvgPool3d((None, 8, 8))  # 输出形状 [B, 512, T, 8, 8]
        )
        
        # 时空注意力模块
        self.st_attention = SpatioTemporalAttention(
            input_dim=8*8*512, 
            embed_dim=1024
        )
        
        # Transformer模块
        self.transformer = nn.Sequential(
            TransformerBlock(1024, num_heads=8),
            TransformerBlock(1024, num_heads=8)
        )
        
        # LSTM时序建模
        self.lstm = nn.LSTM(
            input_size=1024, 
            hidden_size=512,
            bidirectional=True,
            batch_first=True
        )
        
        # 分类头
        self.head = nn.Sequential(
            nn.Linear(1024, 512),
            nn.GELU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        # CNN特征提取
        x = self.cnn(x)  # [B, 512, T, 8, 8]
        
        # 调整维度顺序为 [B, T, 8, 8, 512]
        x = x.permute(0, 2, 3, 4, 1)
        
        # 应用时空注意力
        x = self.st_attention(x)  # 输出形状 [T, B, 1024]
        
        # 调整维度适配后续处理
        x = x.permute(1, 0, 2)  # [B, T, 1024]
        
        # Transformer处理
        x = self.transformer(x)
        
        # LSTM时序建模
        x, _ = self.lstm(x)
        
        # 分类头
        return self.head(x[:, -1, :])

class EnhancedEnsemble:
    def __init__(self, num_models=5, num_classes=6):
        self.num_models = num_models
        self.models = []
        self.optimizers = []
        self.schedulers = []
        self.train_losses = []
        self.val_losses = []
        
        base_lr = 1e-5
        for i in range(num_models):
            model = HARnet(num_classes).to(device)
            lr = base_lr * (10 ** (np.random.uniform(-0.5, 0.5)))
            wd = np.random.choice([0.01, 0.03, 0.05, 0.07])
            
            optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=wd)
            scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.8, patience=12)
            
            self.models.append(model)
            self.optimizers.append(optimizer)
            self.schedulers.append(scheduler)
            
    def train(self, train_loader, val_loader, num_epochs=200):
        scalers = [GradScaler() for _ in range(self.num_models)]
        
        for model_idx in range(self.num_models):
            print(f"\nTraining Model {model_idx+1}/{self.num_models}")
            model_train_losses = []
            model_val_losses = []
            best_loss = float('inf')
            patience_counter = 0

            for epoch in range(num_epochs):
                # 训练阶段
                self.models[model_idx].train()
                train_loss = 0.0
                for batch_idx, (inputs, targets) in enumerate(train_loader):
                    inputs, targets = inputs.to(device), targets.to(device)
                    self.optimizers[model_idx].zero_grad()
                    
                    with autocast():
                        outputs = self.models[model_idx](inputs)
                        loss = criterion(outputs, targets)
                    
                    scalers[model_idx].scale(loss).backward()
                    torch.nn.utils.clip_grad_norm_(self.models[model_idx].parameters(), 1.0)
                    scalers[model_idx].step(self.optimizers[model_idx])
                    scalers[model_idx].update()
                    train_loss += loss.item()
                    
                    if batch_idx % 10 == 0:
                        torch.cuda.empty_cache()

                avg_train_loss = train_loss / len(train_loader)
                model_train_losses.append(avg_train_loss)
                
                # 验证阶段
                val_loss = 0.0
                self.models[model_idx].eval()
                with torch.no_grad():
                    for inputs, targets in val_loader:
                        inputs, targets = inputs.to(device), targets.to(device)
                        outputs = self.models[model_idx](inputs)
                        val_loss += criterion(outputs, targets).item()
                
                avg_val_loss = val_loss / len(val_loader)
                model_val_losses.append(avg_val_loss)
                self.schedulers[model_idx].step(avg_val_loss)
                
                # 早停机制
                if avg_val_loss < best_loss:
                    best_loss = avg_val_loss
                    patience_counter = 0
                    torch.save(
                        self.models[model_idx].state_dict(),
                        f"./saved_models/best_model_{model_idx}.pth"
                    )
                else:
                    patience_counter += 1
                    if patience_counter >= 10:
                        print(f"Model {model_idx} early stopping at epoch {epoch}")
                        break
                
                torch.cuda.empty_cache()
                print(f"Model {model_idx+1} Epoch {epoch+1} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")

            # 保存当前模型的损失记录
            self.train_losses.append(model_train_losses)
            self.val_losses.append(model_val_losses)

# 数据加载部分
def load_dataset(data_dir):
    features = []
    labels = []
    for filename in sorted(os.listdir(data_dir)):
        if filename.endswith('.npz'):
            file_path = os.path.join(data_dir, filename)
            data = np.load(file_path)
            features.extend(data['arr_0'])
            labels.extend(data['arr_1'])
            data.close()
    return np.array(features), np.array(labels)

# 训练配置和初始化
num_epochs = 200
batch_size = 32
save_dir = './saved_models/'
os.makedirs(save_dir, exist_ok=True)

# 加载数据
data_dir = './autodl-tmp/train/' 
features, labels = load_dataset(data_dir)
dataset = HARDataset(features, labels)

# 划分数据集
total = len(dataset)
train_size = int(0.6 * total)
remaining = total - train_size
val_size = remaining // 2
test_size = remaining - val_size

train_dataset, temp_dataset = random_split(dataset, [train_size, remaining])
val_dataset, test_dataset = random_split(temp_dataset, [val_size, test_size])

# 创建数据加载器
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# 损失函数
class_counts = np.bincount([dataset.label_map[label] for label in labels])
weights = 1.0 / (class_counts + 1e-8)
weights = torch.tensor(weights, dtype=torch.float32).to(device)
criterion = nn.CrossEntropyLoss(weight=weights)

# 训练集成模型
ensemble = EnhancedEnsemble(num_models=5)
ensemble.train(train_dataloader, val_dataloader, num_epochs=num_epochs)

# 绘制所有模型的训练曲线
plt.figure(figsize=(20, 12))
for i in range(ensemble.num_models):
    plt.subplot(2, 3, i+1)  # 2行3列布局
    plt.plot(ensemble.train_losses[i], label='Train Loss')
    plt.plot(ensemble.val_losses[i], label='Val Loss')
    plt.title(f'The Training/Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
plt.tight_layout()
plt.savefig('./figures/all_models_training_curves.png', dpi=300, bbox_inches='tight')
plt.show()

# 评估单个模型的函数
def evaluate_single_model(model, test_loader):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            preds = torch.argmax(outputs, dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(targets.numpy())
    return all_preds, all_labels

# 评估所有单个模型
for i in range(ensemble.num_models):
    # 加载模型
    model = HARnet(num_classes=6).to(device)
    model.load_state_dict(torch.load(f"./saved_models/best_model_{i}.pth"))
    
    # 评估
    preds, labels = evaluate_single_model(model, test_dataloader)
    
    # 打印指标
    print(f"\nModel {i+1} Evaluation:")
    print(classification_report(labels, preds, target_names=list(dataset.label_map.keys()), digits=4))
    print(f"Accuracy: {accuracy_score(labels, preds)*100:.2f}%")
    print(f"Weighted F1: {f1_score(labels, preds, average='weighted'):.4f}")
    
    # 绘制混淆矩阵
    cm = confusion_matrix(labels, preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=list(dataset.label_map.keys()))
    disp.plot()
    plt.title(f'The Confusion Matrix')
    plt.savefig(f'./figures/model_{i+1}_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

# 评估集成模型
def evaluate_ensemble(test_loader):
    models = []
    for i in range(5):
        model = HARnet(num_classes=6).to(device)
        model.load_state_dict(torch.load(f"./saved_models/best_model_{i}.pth"))
        model.eval()
        models.append(model)
    
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs = inputs.to(device)
            batch_probs = []
            
            for model in models:
                outputs = model(inputs)
                probs = torch.softmax(outputs, dim=1).cpu().numpy()
                batch_probs.append(probs)
                torch.cuda.empty_cache()
            
            avg_probs = np.mean(batch_probs, axis=0)
            final_preds = np.argmax(avg_probs, axis=1)
            all_preds.extend(final_preds)
            all_labels.extend(targets.numpy())
    
    print("\nEnsemble Model Evaluation:")
    print(classification_report(all_labels, all_preds, target_names=list(dataset.label_map.keys()), digits=4))
    print(f"Accuracy: {accuracy_score(all_labels, all_preds)*100:.2f}%")
    print(f"Weighted F1: {f1_score(all_labels, all_preds, average='weighted'):.4f}")
    
    cm = confusion_matrix(all_labels, all_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=list(dataset.label_map.keys()))
    disp.plot()
    plt.title('Ensemble Model Confusion Matrix')
    plt.savefig('./figures/ensemble_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

# 执行集成评估
evaluate_ensemble(test_dataloader)
```
